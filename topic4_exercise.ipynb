## Exercise 3: The Batch Norm Bias Trap

A computation block is given below. The code runs perfectly and gets great accuracy. But, whoever wrote it didn't read Chapter 9.3, and they left a tiny, invisible memory leak in the architecture. They set use_bias=True on every Convolutional layer.

Your Task:

1. Run the code below as-is. Look at the Total Params at the bottom. Go through the code and change every use_bias=True to use_bias=False.

2. Run the cell again and look at the new Total Params.

You just deleted several hundred parameters, saving memory, while keeping the model performance similar(within +/-5%).

3. Go to the Mentiboard on the Presenter and solve the MCQ there. Fastest correct answer gets the most points.

```python
def build_model(bias_flag):
    """A factory function to easily build our model with or without biases!"""
    inputs = keras.Input(shape=(28, 28, 1))
    x = layers.Rescaling(1./255)(inputs)

    # We pass the bias_flag to every layer
    x = layers.Conv2D(filters=32, kernel_size=3, padding="same", use_bias=bias_flag)(x)
    residual = x

    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.SeparableConv2D(64, 3, padding="same", use_bias=bias_flag)(x)

    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    x = layers.SeparableConv2D(64, 3, padding="same", use_bias=bias_flag)(x)

    x = layers.MaxPooling2D(3, strides=2, padding="same")(x)
    residual = layers.Conv2D(64, 1, strides=2, padding="same", use_bias=bias_flag)(residual)

    x = layers.add([x, residual])
    x = layers.Flatten()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(10, activation="softmax")(x)

    return keras.Model(inputs=inputs, outputs=outputs)

# ---------------------------------------------------------
# THE A/B TEST (Optimized for Live Execution)
# ---------------------------------------------------------
# To save time, we will only train on a
# random subset of 5,000 images rather than all 60,000.

# 1. The Unoptimized Model (Biases = True)
print("--- MODEL A: (BIASES = TRUE) ---")
model_a = build_model(bias_flag=True)
print(f"Total Parameters: {model_a.count_params():,}")
model_a.compile(optimizer="rmsprop", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Train on a 5000-image slice for blazing fast execution!
model_a.fit(train_images[:5000], train_labels[:5000], epochs=1, batch_size=256, validation_split=0.2)


# 2. Your Optimized Model
print("\n--- MODEL B: (OPTIMIZED) ---")
# TODO: Change this flag to False!
model_b = build_model(bias_flag=True)

print(f"Total Parameters: {model_b.count_params():,}")
model_b.compile(optimizer="rmsprop", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

# Train on the exact same 5000-image slice to compare
model_b.fit(train_images[:5000], train_labels[:5000], epochs=1, batch_size=256, validation_split=0.2)
```
